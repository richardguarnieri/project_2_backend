```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
from config import pg_pass
```

```python
movie_file = "movies_info.csv"
movies_info_df = pd.read_csv(movie_file, index_col=0)
movies_info_df.head()
```

```python
rate_file = "ratings.csv"
rate_info_df = pd.read_csv(rate_file, index_col=0)
rate_info_df.head()
```

```python
move_b_file = "movies_info_b.csv"
movies_b_df = pd.read_csv(move_b_file, index_col=0)
movies_b_df.head()
```

```python
rds_connection_string = (f'postgres:{pg_pass}@localhost:5432/ETL-Team4')
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
movies_info_df.to_sql(name='imdb_movies', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from imdb_movies', con=engine).head()
```

```python
rate_info_df.to_sql(name='imdb_ratings', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from imdb_ratings', con=engine).head()
```

```python
movies_b_df.to_sql(name='imdb_movies_b', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from imdb_movies_b', con=engine).head()
```
